## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns





# Create a new DataFrame 'y_df' containing only 'Attrition' and 'Department' columns
y_df = attrition_df[['Attrition', 'Department']]
# Display the first 5 rows to verify the result
y_df.head()


```

```python
# Create a list of at least 10 column names to use as X data
x_columns = ['Age', 'JobSatisfaction', 'OverTime', 'DistanceFromHome',
             'Education', 'NumCompaniesWorked', 'EnvironmentSatisfaction',
             'HourlyRate', 'JobInvolvement', 'PerformanceRating']

# Create X_df using your selected columns
X_df = attrition_df[x_columns]

# Show the data types for X_df

print(X_df.dtypes)



```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split



# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)

# Verify the shape of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

from sklearn.preprocessing import OneHotEncoder

# Convert training data to numeric types
OverTime_encoder = OneHotEncoder (sparse_output=False, drop='first', handle_unknown='ignore')
OverTime_encoder.fit(np.array(X_train['OverTime']).reshape(-1, 1))
X_train ['OverTimeEncoder'] = OverTime_encoder.transform(np.array(X_train['OverTime']).reshape(-1, 1))
X_test ['OverTimeEncoder'] = OverTime_encoder.transform(np.array(X_test['OverTime']).reshape(-1, 1))


X_train = X_train.drop(columns=['OverTime'])
X_test = X_test.drop(columns=['OverTime'])


print(X_train.dtypes)
print(X_test.dtypes)

```

```python
# Create a StandardScaler


# Fit the StandardScaler to the training data


# Scale the training and testing data






from sklearn.preprocessing import StandardScaler

# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Verify the shapes of the scaled data
print("X_train_scaled shape:", X_train_scaled.shape)
print("X_test_scaled shape:", X_test_scaled.shape)


```

```python
# Create a OneHotEncoder for the Department column


# Fit the encoder to the training data


# Create two new variables by applying the encoder
# to the training and testing data





from sklearn.preprocessing import OneHotEncoder

# Convert training data to numeric types
Attrition_encoder = OneHotEncoder (sparse_output=False, drop= None, handle_unknown='ignore')
Attrition_encoder.fit(np.array(y_train['Attrition']).reshape(-1, 1))
y_attrition_train = Attrition_encoder.transform(np.array(y_train['Attrition']).reshape(-1, 1))
y_attrition_test = Attrition_encoder.transform(np.array(y_test['Attrition']).reshape(-1, 1))


display(y_attrition_train)
display(y_attrition_test)




```

```python
# Create a OneHotEncoder for the Attrition column


# Fit the encoder to the training data


# Create two new variables by applying the encoder
# to the training and testing data




from sklearn.preprocessing import OneHotEncoder

# Convert training data to numeric types
department_encoder = OneHotEncoder (sparse_output=False, drop= None, handle_unknown='ignore')
department_encoder.fit(np.array(y_train['Department']).reshape(-1, 1))
y_department_train = department_encoder.transform(np.array(y_train['Department']).reshape(-1, 1))
y_department_test = department_encoder.transform(np.array(y_test['Department']).reshape(-1, 1))


display(y_department_train)
display(y_department_test)





```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
number_input_features = len(X_train.columns)
number_input_features

# Create the input layer
input = layers.Input(shape=(number_input_features,))

# Create at least two shared layers
share1 = layers.Dense(units=64, activation='relu')(input)
share2 = layers.Dense(units=128, activation='relu')(share1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(units=64, activation='relu', name='department_hidden')(share2)


# Create the output layer
department_output = layers.Dense(y_department_train.shape[1], activation='sigmoid' , name='department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(units=64, activation='relu', name='attrition_hidden')(share2)
# Create the output layer
attrition_output = layers.Dense(y_attrition_train.shape[1], activation='sigmoid', name='attrition_output')(attrition_hidden)
```

```python
# Create the model
model= Model(inputs=input, outputs=[attrition_output, department_output])

# Compile the model
model.compile(optimizer='adam', loss={'attrition_output': 'binary_crossentropy', 'department_output': 'categorical_crossentropy'}, 
              metrics={'attrition_output': 'accuracy', 'department_output': 'accuracy'})



# Summarize the model
model.summary()

```

```python
# Train the model

fit_model = model.fit(X_train_scaled,  {'department_output': y_department_train, 'attrition_output': y_attrition_train}, epochs=100, verbose=1)






```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled, {'department_output': y_department_test , 'attrition_output': y_attrition_test})

```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {test_results[3]*100:.2f}%")
print(f"Attrition Accuracy: {test_results[4]*100:.2f}%")
```
